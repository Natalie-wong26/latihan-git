### Conclusion from Data Info & Actions Follow:
- Dataset comprise of Master Data and Transactional Data. Personally, I would prefer separating master data and transactional data. But since I do not plan to import the dataset into a database and I would end up 'joining' the data anyway, i proceed with this dataset as is.
- `Location` has 34 unique values which is 33 Indonesian provinces + Indonesia (Country) and `Location ISO Code` has the same information but in codes based on ISO standards. So, I decide to drop `Location ISO Code`.
- Indonesia has 34 provinces, so it means there is one province not included in the dataset. Which is Gorontalo in Sulawesi Island.
- I will also Filtered out rows that contain Location == 'Indonesia'
- `Location Level` indicate wheter the data is collected at province or country level. Since I filtered out row with country level data (Location == 'Indonesia') all data will be in province level and this column will be drop
- Now `Location` and `Province` has the same value. I will drop `Location` column after i filter the data.
- `City or Regency`, `Name`, `Item`, `Kind` and `Hidden` columns have no value (all missing) and since there is no significant use for these columns. I decide to drop these columns.  
- Since `City or Regency` values don't exist in this dataset, this analysis will be done in province, island, and country level.  
- `Total Cases`, `Total Deaths`, `Total Recovered`, and `Total Active Cases` refer total cases until related date (`Date` column). These columns will be drop.
- `Longitude` and `Latitude` will be kept for visualization with Power BI

- I decided the column that I would use are the following: `Date`, `New Cases`, `New Deaths`, `New Recovered`, `New Active Cases`, `Province`, `Island`, `Area (km2)`, `Population`, `Population Density`, `Longitude` and `Latitude`

```python
#Filter out rows with location == 'Indonesia'
data_edited = data.loc[(data['Location'] != 'Indonesia')]
data_edited.head(3)
```

```python
#copy
data_edited = data_edited[['Date','Island', 'Location ISO Code', 'Province', 'Country', 'New Cases', 'New Deaths', 'New Recovered', 'New Active Cases', 
                           'Area (km2)', 'Population', 'Population Density', 'Longitude', 'Latitude']]
data_edited.tail()
```

## Replace Space in Column Name with Underscore(_)

```python
data_edited.columns = data_edited.columns.str.replace(' ', '_')
data_edited.tail()
```
